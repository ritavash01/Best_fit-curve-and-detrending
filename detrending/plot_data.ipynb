Plot for original simulated data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = "/home/ritavash/Documents/Simulated_Data/simulated_RFI_1.csv"  
magnitude = np.loadtxt(file_path, delimiter=",")

# Parameters
time_points = magnitude.shape[1]  # Get time_points from the shape of the matrix
frequency_points = magnitude.shape[0]  # Get frequency_points from the shape of the matrix

# Optionally, you can visualize the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(magnitude, aspect='auto', origin='lower', cmap='viridis', extent=[0, time_points, 0, frequency_points])
plt.colorbar(label='Magnitude')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Heatmap ')
plt.show()
```
